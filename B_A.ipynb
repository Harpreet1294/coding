```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('./BA_Project.csv', encoding= 'unicode_escape')
df.head()
```

```
columns = [i.replace('\xa0','') for i in df.columns]
```

```
del columns[:4]
```

```
df = df.replace(r'^\s*$', 0, regex=True)
df.head()
```

```
df = df.rename(columns=lambda x: x.strip())
```

```
for col in columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
```

```
for col in columns:
    df[col] = df[col].apply(pd.to_numeric)
df.head()
```

```
Theory = columns[:-5]
```

```
for col in Theory:
    df1 = df[(df[col] < 75)]
```

```
df1.head()
```

```
Practical = columns[-5:-1]
```

```
for col in Practical:
    df2 = df[(df[col] < 75)]
```

```
df2.head()
```

```
df1.to_csv('ListTheory.csv')
```

```
df1.head()
```

```
df2.to_csv('ListPractical.csv')
```

```
df3 = df[~df.index.isin(df1.index)]
```

```
df_new = df3[~df3.index.isin(df2.index)]
```

```
df_new.head()
```

```
df_new.to_csv('Attendance.csv')
```
